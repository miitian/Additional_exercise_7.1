```python
import sklearn as sl
import pandas as pd
import seaborn as sns

from sklearn.datasets import load_iris
```

```python
#Import Iris data set
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
df.head()
```

```python
label= pd.DataFrame(list(map(lambda x: iris.target_names[x], iris.target)), columns =['Species'])
```

```python
label.head()
```

```python
iris = pd.concat([df, label], axis=1)
```

```python
iris.head()
```

```python
df1 = iris.drop('Species', axis=1)
#df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#fig = plt.figure(figsize=(14,6))
#plt.figure(figsize=(14,6))
fig, ax = plt.subplots(2, 2,figsize=(14,6))


sns.distplot(df1['sepal length (cm)'], label='sepal length', hist= False, ax=ax[0,0], color='r')
sns.distplot(df1['sepal width (cm)'], label='sepal width', hist= False, ax=ax[0,1], color='g')
sns.distplot(df1['petal length (cm)'], label='petal length', hist= False, ax=ax[1,0], color='b')
sns.distplot(df1['petal width (cm)'], label='petal width', hist= False, ax=ax[1,1], color='orange')

plt.legend()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(data=df1)
```

```python
plt.figure(figsize=(14,6))
sns.countplot(iris.Species)
```

```python
plt.figure(figsize=(14,6))
sns.pairplot(data=iris)
```

```python
sns.lmplot(data=iris, x='sepal length (cm)', y='petal length (cm)', hue='Species')
```

```python
sns.barplot(data=iris, x= 'Species', y='sepal length (cm)')
```
